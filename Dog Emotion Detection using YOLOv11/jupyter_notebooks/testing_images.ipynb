```python
from ultralytics import YOLO
import cv2
import os

def process_images(model_path, input_folder, output_folder):
    model = YOLO(model_path)

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            img_path = os.path.join(input_folder, filename)
            image = cv2.imread(img_path)

            results = model(image)
            annotated_image = results[0].plot()

            output_path = os.path.join(output_folder, filename)
            cv2.imwrite(output_path, annotated_image)
            print(f"Saved: {output_path}")

    print(" All images processed.")

# Update these paths
model_path = r"C:\Users\Abhijeet Singh\Downloads\tensorflow\data_augmented1\Data\runs\detect\train\weights\best.pt"
input_folder = r"C:\Users\Abhijeet Singh\Downloads\tensorflow\data_reduced002\annotated_data\test\images"
output_folder = r"C:\Users\Abhijeet Singh\Downloads\tensorflow\data_augmented1\Data\runs\detect\predict"

process_images(model_path, input_folder, output_folder)
```
